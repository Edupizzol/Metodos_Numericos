# Questão 4 — Diagrama de Bifurcação do Mapa Logístico

Mapa logístico:
$$
G(x) = r\,x\,(1 - x)
$$

Serão gerados diagramas de bifurcação para:
- $r \in [2{,}5,\; 4{,}0]$
- $r \in [3{,}62,\; 3{,}64]$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Implementação do mapa logístico e do diagrama de bifurcação

```python
def mapa_logistico(x, r):
    """Aplica uma iteração do mapa logístico G(x) = r*x*(1 - x)."""
    return r * x * (1 - x)


def diagrama_bifurcacao(r_min, r_max, n_r=2000,
                         n_transiente=500, n_pontos=300,
                         x0=0.5):
    """
    Calcula os pontos do diagrama de bifurcação do mapa logístico.

    Parâmetros
    ----------
    r_min, r_max : float
        Intervalo do parâmetro r.
    n_r : int
        Número de valores de r amostrados.
    n_transiente : int
        Iterações descartadas (transiente).
    n_pontos : int
        Iterações coletadas após o transiente.
    x0 : float
        Condição inicial.

    Retorna
    -------
    r_vals : ndarray
        Valores de r repetidos para cada ponto coletado.
    x_vals : ndarray
        Valores de x no regime assintótico.
    """
    r_array = np.linspace(r_min, r_max, n_r)
    r_vals = []
    x_vals = []

    for r in r_array:
        x = x0
        # Descarta transiente
        for _ in range(n_transiente):
            x = mapa_logistico(x, r)
        # Coleta regime assintótico
        for _ in range(n_pontos):
            x = mapa_logistico(x, r)
            r_vals.append(r)
            x_vals.append(x)

    return np.array(r_vals), np.array(x_vals)
```

## (a1) Diagrama de bifurcação para r ∈ [2.5, 4.0]

```python
r_vals, x_vals = diagrama_bifurcacao(r_min=2.5, r_max=4.0, n_r=3000)

plt.figure(figsize=(10, 5))
plt.plot(r_vals, x_vals, ',k', markersize=0.3, alpha=0.5)
plt.xlabel('r')
plt.ylabel('x')
plt.title('Diagrama de Bifurcação — Mapa Logístico  $G(x) = rx(1-x)$,  $r \\in [2.5, 4.0]$')
plt.tight_layout()
plt.show()
```

## (a2) Diagrama de bifurcação para r ∈ [3.62, 3.64]

```python
r_vals2, x_vals2 = diagrama_bifurcacao(r_min=3.62, r_max=3.64,
                                        n_r=2000, n_transiente=1000, n_pontos=500)

plt.figure(figsize=(10, 5))
plt.plot(r_vals2, x_vals2, ',k', markersize=0.5, alpha=0.5)
plt.xlabel('r')
plt.ylabel('x')
plt.title('Diagrama de Bifurcação — Mapa Logístico  $G(x) = rx(1-x)$,  $r \\in [3.62, 3.64]$')
plt.tight_layout()
plt.show()
```
