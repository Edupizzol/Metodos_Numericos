# Questão 1 — Eliminação de Gauss com Pivotamento Parcial

Sistema de equações:

$$
-3x_2 + 7x_3 = 2
$$
$$
x_1 + 2x_2 - x_3 = 3
$$
$$
5x_1 - 2x_2 = 2
$$

```python
import numpy as np
```

## Implementação do método

```python
def gauss_pivotamento_parcial(A, b):
    """
    Resolve o sistema Ax = b usando eliminação de Gauss com pivotamento parcial.

    Parâmetros
    ----------
    A : array_like, shape (n, n)
        Matriz dos coeficientes.
    b : array_like, shape (n,)
        Vetor dos termos independentes.

    Retorna
    -------
    x : ndarray, shape (n,)
        Vetor solução.
    """
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    # Matriz aumentada [A | b]
    Ab = np.hstack([A, b.reshape(-1, 1)])

    # Eliminação progressiva com pivotamento parcial
    for k in range(n - 1):
        # Encontra o índice do maior pivô em valor absoluto abaixo da linha k
        max_idx = np.argmax(np.abs(Ab[k:, k])) + k

        # Troca as linhas k e max_idx
        if max_idx != k:
            Ab[[k, max_idx]] = Ab[[max_idx, k]]

        if Ab[k, k] == 0:
            raise ValueError("Sistema singular: pivô nulo encontrado.")

        # Eliminação
        for i in range(k + 1, n):
            fator = Ab[i, k] / Ab[k, k]
            Ab[i, k:] -= fator * Ab[k, k:]

    # Substituição retroativa
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (Ab[i, -1] - np.dot(Ab[i, i + 1:n], x[i + 1:n])) / Ab[i, i]

    return x
```

## Dados do problema

```python
# Matriz dos coeficientes
#   Equação 1:  0*x1 - 3*x2 + 7*x3 = 2
#   Equação 2:  1*x1 + 2*x2 - 1*x3 = 3
#   Equação 3:  5*x1 - 2*x2 + 0*x3 = 2
A = [
    [ 0, -3,  7],
    [ 1,  2, -1],
    [ 5, -2,  0]
]

b = [2, 3, 2]
```

## Resolução e resultado

```python
x = gauss_pivotamento_parcial(A, b)

print("Solução do sistema:")
for i, xi in enumerate(x, start=1):
    print(f"  x{i} = {xi:.6f}")

# Verificação: resíduo Ax - b
residuo = np.dot(A, x) - np.array(b, dtype=float)
print(f"\nResíduo (Ax - b): {residuo}")
```
