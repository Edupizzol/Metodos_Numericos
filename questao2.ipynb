# Questão 2 — Corrente Eficaz (RMS) por Integração Numérica

Corrente instantânea:
$$
i(t) = 5\,e^{-1{,}25t}\,\mathrm{sen}(2\pi t), \quad T = 1\,\mathrm{s}
$$

Corrente eficaz:
$$
I_{\mathrm{RMS}} = \sqrt{\frac{1}{T}\int_0^T [i(t)]^2\,dt}
$$

Será calculada sobre **meio período** $t \in [0,\, T/2]$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## (a) Gráfico de i(t) no intervalo [0, T/2]

```python
def corrente(t):
    return 5 * np.exp(-1.25 * t) * np.sin(2 * np.pi * t)

T = 1.0  # período (s)

t_plot = np.linspace(0, T / 2, 500)
i_plot = corrente(t_plot)

plt.figure(figsize=(8, 4))
plt.plot(t_plot, i_plot, color='steelblue', linewidth=2)
plt.xlabel('Tempo (s)')
plt.ylabel('Corrente i(t) (A)')
plt.title(r'$i(t) = 5\,e^{-1{,}25t}\,\mathrm{sen}(2\pi t)$ — Meio período')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Implementação da Regra 1/3 de Simpson

```python
def simpson_13(f, a, b, n):
    """
    Integra f no intervalo [a, b] usando a Regra 1/3 de Simpson composta.

    Parâmetros
    ----------
    f : callable
        Função a integrar.
    a : float
        Limite inferior de integração.
    b : float
        Limite superior de integração.
    n : int
        Número de segmentos (deve ser par).

    Retorna
    -------
    integral : float
        Valor aproximado da integral.
    """
    if n % 2 != 0:
        raise ValueError("O número de segmentos n deve ser par para a regra 1/3 de Simpson.")

    h = (b - a) / n
    t = np.linspace(a, b, n + 1)
    y = f(t)

    # Coeficientes: 1, 4, 2, 4, 2, ..., 4, 1
    coef = np.ones(n + 1)
    coef[1:-1:2] = 4  # índices ímpares
    coef[2:-2:2] = 2  # índices pares internos

    integral = (h / 3) * np.dot(coef, y)
    return integral
```

## (b) Cálculo da corrente eficaz I_RMS no meio período

```python
# Parâmetros
T = 1.0          # período (s)
a = 0.0          # início do intervalo
b = T / 2        # meio período
n_seg = 32       # número de segmentos (par)

# Integrando: [i(t)]^2
def integrando(t):
    return corrente(t) ** 2

# Integral de [i(t)]^2 de 0 a T/2
integral = simpson_13(integrando, a, b, n_seg)

# Corrente eficaz: sqrt((1/T) * integral)
I_RMS = np.sqrt((1 / T) * integral)

print(f"Integral de [i(t)]^2 de 0 a T/2 : {integral:.6f} A²·s")
print(f"Corrente eficaz I_RMS            : {I_RMS:.6f} A")
```
